### numpy, panda, matplotlib练习

```python
%matplotlib inline
"""导入所需要的包"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(0.0, 5.0) #产生数据
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
plt.subplot(2,1,1) #绘制第一个子图
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2 ,y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```
